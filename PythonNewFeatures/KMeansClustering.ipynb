```python
df = merged_df
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = MinMaxScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns = x.columns)
x
```

```python
def IDPsInCluster(cluster):
    return (cluster["label"] == 1).sum()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

SSERandom = []
SSEKmeans = []
silhouette_scoresRandom = []
silhouette_scoresKmeans = []

numOfclusters = range(100, 801, 50)
for i in numOfclusters:
    for init in ['k-means++', 'random']:
        print(f"{i} clusters, init : {init}")

        kmeans=KMeans(n_clusters=i, init=init, n_init = 10)
        kmeans.fit(x)

        if init == 'random':
            SSERandom.append(kmeans.inertia_)
            silhouette_scoresRandom.append(silhouette_score(x, kmeans.labels_))
        else:
            SSEKmeans.append(kmeans.inertia_)
            silhouette_scoresKmeans.append(silhouette_score(x, kmeans.labels_))

        for j in range(0,i):
            cluster= df[kmeans.labels_ == j]
            #print(cluster["node"])
            print(f"\t{cluster.shape[0]} instances in cluster {j}.")
            print(f"\t\t{IDPsInCluster(cluster)} IDPs and {cluster.shape[0] - IDPsInCluster(cluster)} non IDPs.")
        print("-"*50)
```

```python
def add_cell(figure, subplot_size, plot_ind, x, y, title, y_label):
    fig.add_subplot(subplot_size[0], subplot_size[1], plot_ind)
    plt.plot(x, y, marker = 'o')
    plt.xticks(x)
    plt.title(title)
    plt.xlabel("Broj klastera")
    plt.ylabel(y_label)
```

```python
fig = plt.figure(figsize=(10,10))

add_cell(fig, (2, 2), 1, numOfclusters, SSERandom, "SSE Random", "SEE")
add_cell(fig, (2, 2), 2, numOfclusters, SSEKmeans, "SSE k-means++", "SEE")
add_cell(fig, (2, 2), 3, numOfclusters, silhouette_scoresRandom, "Silhouette scores Random", "Silhouette score")
add_cell(fig, (2, 2), 4, numOfclusters, silhouette_scoresKmeans, "Silhouette scores k-means++", "Silhouette score")

plt.tight_layout()
plt.show()
```
